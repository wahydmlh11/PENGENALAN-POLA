```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

# Muat dataset Diabetes
dataset_diabetes = load_diabetes()

# Tetapkan fitur (X) dan target (y)
X = dataset_diabetes.data
y = dataset_diabetes.target

# Bagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Inisialisasi model KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=1)

# Latih model
knn.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred = knn.predict(X_test)

# Cetak hasil prediksi pada data uji
print("Prediksi pada dataset uji:\n{}".format(y_pred))

```

```python
from sklearn.metrics import mean_squared_error

# Menghitung skor model pada data uji
score = knn.score(X_test, y_test)

# Menghitung akurasi model pada data uji
accuracy = 1 - mean_squared_error(y_test, y_pred)

# Cetak skor (akurasi) model pada data uji
print("Skor pada dataset uji: {:.2f}".format(score))
print("Akurasi pada dataset uji: {:.2f}".format(accuracy))
```
