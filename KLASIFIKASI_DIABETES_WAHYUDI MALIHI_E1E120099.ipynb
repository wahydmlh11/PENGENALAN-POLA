```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# Inisialisasi model Random Forest
model = RandomForestClassifier()

# Memasukkan data fitur dari pengguna
pregnancies = int(input("Jumlah kehamilan: "))
glucose = float(input("Kadar glukosa: "))
blood_pressure = float(input("Tekanan darah: "))
skin_thickness = float(input("Ketebalan kulit: "))
insulin = float(input("Kadar insulin: "))
bmi = float(input("Indeks massa tubuh (BMI): "))
diabetes_pedigree = float(input("Diabetes pedigree function: "))
age = float(input("Usia: "))
# Additional features
total_serum_insulin = float(input("Total serum insulin: "))
diabetes_pedigree_2 = float(input("Diabetes pedigree function 2: "))

# Menyiapkan data fitur pengguna sebagai array numpy
user_data = np.array([[pregnancies, glucose, blood_pressure, skin_thickness, insulin, bmi, diabetes_pedigree, age, total_serum_insulin, diabetes_pedigree_2]])

# Memuat dataset diabetes
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()

# Memisahkan fitur dan label dari dataset
X = diabetes.data
y = diabetes.target

# Melatih model dengan seluruh data diabetes
model.fit(X, y)

# Mengklasifikasikan data pengguna
prediction = model.predict(user_data)

# Menampilkan hasil klasifikasi
if prediction[0] == 0:
    print("Pasien tidak menderita diabetes.")
else:
    print("Pasien menderita diabetes.")
1
```
